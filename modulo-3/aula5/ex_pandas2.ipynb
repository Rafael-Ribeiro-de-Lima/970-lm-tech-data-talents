```python
import os
import pandas as pd
```

```python
df = pd.read_csv(r'db\avocado_with_type.csv')
df.head()
```

```python
# Quantidade de abacates do tipo convecional e do tipo orgânico.

df['type'].value_counts()
```

```python
# Quantidade de registros (linhas) de abacates vendidos para cada região.

df['region'].value_counts()
```

```python
# Renomeia aqueles que estão iniciando em letra minúscula para que iniciem em letra maiúscula.
df = df.rename(columns=lambda coluna: coluna.capitalize())
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
# Remove a coluna que apresenta o ano.

df = df.drop(columns='Year')
df.head()
```
